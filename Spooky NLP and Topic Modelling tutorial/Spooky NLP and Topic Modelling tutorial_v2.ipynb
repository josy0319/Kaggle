# 2.Natural Language Processing

> **Text pre-processing can be boiled down to these few simple steps:**
- 1.Tokenization - Segregation of the text into its individual constituent words
- 2.Stopwords -Throw away any words that occur too frequently as its frequency of occurrence will not be useful in helping detecting relevant texts = 불용어 = 의미가 없는 단어 = a, to, am 등
- 3.Stemming - combine variants of words into a single parent word that still conveys the same meaning = 단어를 원형으로 표현 = Knowing, Known => know
- 4.Vectorization - Converting text into vector format. One of the simplest is the famous bag-of-words approach, where you create a matrix

# 3.Topic modeling

```python
def print_top_words(model, feature_names, n_top_words):
    for index, topic in enumerate(model.components_):
        message = "\nTopic #{}:".format(index)
        message += " ".join([feature_names[i] for i in topic.argsort()[:-n_top_words - 1 :-1]])
        print(message)
        print("="*70)
#lda 모델을 사용하여 top words를 출력하는 함수
```

**3a. Putting all the preprocessing steps together**

```python
lemm = WordNetLemmatizer()
class LemmaCountVectorizer(CountVectorizer):
    def build_analyzer(self):
        analyzer = super(LemmaCountVectorizer, self).build_analyzer()
        return lambda doc: (lemm.lemmatize(w) for w in analyzer(doc))
#Extending the CountVectorizer class with a lemmatizer
#CountVectorizer란 단어 수를 count해서 vector로 만들어주는 모듈
#출처 : https://blog.naver.com/myincizor/221643794025
```

```python
text = list(train.text.values)
tf_vectorizer = LemmaCountVectorizer(max_df=0.95,
                                    min_df=2,
                                    stop_words='english',
                                    decode_error='ignore')
tf = tf_vectorizer.fit_transform(text)
#tf = <19579x13781 sparse matrix of type '<class 'numpy.int64'>' with 212037 stored elements in Compressed Sparse Row format>
```

```python
feature_names = tf_vectorizer.get_feature_names()
#벡터화시킨 단어들의 이름들을 저장
count_vec = np.asarray(tf.sum(axis=0)).ravel()
#벡터화 된 array를 axis=0방향으로 더한 후 저장 -> 등장한 단어 빈도 수
zipped = list(zip(feature_names, count_vec))
x, y = (list(x) for x in zip(*sorted(zipped, key=lambda x:x[1], reverse=True)))
#x,y 13781개 -> 빈도수가 높은 순서로 정렬 후 단어와 빈도 수 나눠서 저장

Y = np.concatenate([y[0:15], y[-16:-1]])
X = np.concatenate([x[0:15], x[-16:-1]])
#빈도 수가 높고 낮은 15개를 합친 array 생성

data = [go.Bar(
            x = x[0:50],
            y = y[0:50],
            marker = dict(colorscale='Jet',
                         color = y[0:50]),
                         text='Word counts')]
layout = go.Layout(title='Top 50 word frequencies after Preprocessing')
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='basic-bar')

data = [go.Bar(x=x[-100:],
               y=y[-100:],
               marker=dict(colorscale='Portland',
                           color=y[-100]),
                           text='Word counts')]
layout = go.Layout(title='Bottom 100 Word frequencies after Preprocessing')
fig = go.Figure(data=data, layout=layout)
py.iplot(fig, filename='basic-bar')
```

**3b. Latent Dirichlet Allocation(LDA)**

```python
lda = LatentDirichletAllocation(n_components=11, max_iter=5,
                               learning_method='online',
                               learning_offset=50.,
                               random_state=0)
#토픽모델을 분류하는 모델
#문서의 의미에 영향을 끼치치 않는 he,she,a같은 요소를 제거하고 토픽 모델을 만들 수 있음
```

```python
lda.fit(tf)
```

- Topics generated by LDA

```python
n_top_words = 40
print('\nTopics in LDA model : ')
tf_feature_names = tf_vectorizer.get_feature_names()
print_top_words(lda, tf_feature_names, n_top_words)
#LDA모델을 거친 상위 40개의 Topic을 출력
```

```python
first_topic = lda.components_[0]
second_topic = lda.components_[1]
third_topic = lda.components_[2]
fourth_topic = lda.components_[3]
```

```python
first_topic_words = [tf_feature_names[i] for i in first_topic.argsort()[:-50 - 1 :-1]]
second_topic_words = [tf_feature_names[i] for i in second_topic.argsort()[:-50 - 1 :-1]]
third_topic_words = [tf_feature_names[i] for i in third_topic.argsort()[:-50 - 1 :-1]]
fourth_topic_words = [tf_feature_names[i] for i in fourth_topic.argsort()[:-50 - 1 :-1]]
#argsort로 데이터의 빈도 순서대로 인덱스를 매긴 후 역으로 출력
#즉, 많이 lda모델 평가 값이 높은 순으로 출력
```
